# Clustering

The task of this project is to cluster two data sets as well as possible. The first dataset contains gene expression data while the second is about mass spectrometry. The goodness of the clustering is evaluated with normalized mutual information (NMI). In this notebook, the gene expression clustering is performed.

## Data 1: Gene expression

### Clustering

In this section, different clustering methods are performed.

#### K-Means

Analyzing the optimal number of clusters using Silhouette coefficent, Calinski Harabasz coefficent, Davies-Bouldin coefficent, and the elbow method.

```python
silhouette=[]
calinski=[]
davies=[]
K=[2, 3, 4, 5, 6, 7, 10,20]
for i in K:
    kmeans_model = KMeans(n_clusters=i)
    kmeans_model = kmeans_model.fit(df_gene_cleaned)
    labels = kmeans_model.labels_
    silhouette.append(metrics.silhouette_score(df_gene_cleaned, labels))
    calinski.append(metrics.calinski_harabasz_score(df_gene_cleaned, labels))
    davies.append(metrics.davies_bouldin_score(df_gene_cleaned, labels))

fig, axs = plt.subplots(3, figsize=(16,12))
axs[0].plot(K, silhouette)
axs[1].plot(K, calinski)
axs[2].plot(K, davies)
axs[0].title.set_text('Silhouette Coefficient')
axs[1].title.set_text('Calinski Harabasz Coefficient')
axs[2].title.set_text('Davies-Bouldin Coefficient')
plt.show()
```

```python
distortions = []
K = range(1,20)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(df_gene_cleaned)
    distortions.append(kmeanModel.inertia_)

plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
plt.show()
```

The optimal number of clusters seems to be around 5. However, also other values shall be tested. Reasoning:
* In the Elbow Method, the optimal k is at the elbow, which seems to be at the position k=5.
* Sihuette: The best value is 1 and the worst value is -1. Values near 0 indicate overlapping clusters. => Value is greatest with k=6 or k=7.
* Calinski Harabasz: Higher value is better. The maximum is at k=2, and decreasing from there.
* Davies-Bouldin: The minimum score is zero, with lower values indicating better clustering. => Minimum at k=2, and increasing from there.

##### Visualizing different k-values using PCA

```python
clusterer = KMeans(n_clusters=2)
labels2 = clusterer.fit_predict(df_gene_cleaned)
clusterer = KMeans(n_clusters=5)
labels5 = clusterer.fit_predict(df_gene_cleaned)
clusterer = KMeans(n_clusters=10)
labels10 = clusterer.fit_predict(df_gene_cleaned)
    
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(df_gene_cleaned)
fig, axes = plt.subplots(1, 3, figsize=(20,8))
axes[0].set_xlabel('Principal Component 1', fontsize = 15)
axes[0].set_ylabel('Principal Component 2', fontsize = 15)
axes[0].set_title('2 component PCA', fontsize = 20)
axes[0].scatter(principalComponents[:,0], principalComponents[:,1],c=labels2, s=50)
axes[0].grid()
axes[1].set_xlabel('Principal Component 1', fontsize = 15)
axes[1].set_ylabel('Principal Component 2', fontsize = 15)
axes[1].set_title('5 component PCA', fontsize = 20)
axes[1].scatter(principalComponents[:,0], principalComponents[:,1],c=labels5, s=50)
axes[1].grid()
axes[2].set_xlabel('Principal Component 1', fontsize = 15)
axes[2].set_ylabel('Principal Component 2', fontsize = 15)
axes[2].set_title('10 component PCA', fontsize = 20)
axes[2].scatter(principalComponents[:,0], principalComponents[:,1],c=labels10, s=50)
axes[2].grid()
```

##### K-means NMI

```python
K = range(2,10)
for k in K:
    clusterer = KMeans(n_clusters=k)
    labels = clusterer.fit_predict(df_gene_cleaned)
    nmi = metrics.normalized_mutual_info_score(df_gene["class"],labels, average_method="geometric")
    print("NMI for K-means with {} clusters: {}".format(k, round(nmi,2)))
```

It turned out that with K-means clustering the best NMI achieved was 0.88 with 6 clusters.

#### Hierarchical agglomerative clustering

In this section, hierarchical agglomerative clustering is performed using the following linkage metrics:
* single-linkage metric
* complete-linkage metric
* average-linkage metric
* ward-linkage metric

```python
K = range(2,20)
for l in ["single", "average", "complete", "ward"]:
    nmis={}
    for k in K:
        clustering = AgglomerativeClustering(n_clusters=k, affinity='euclidean', compute_full_tree='auto', linkage=l)
        clusters=clustering.fit_predict(df_gene_cleaned)
        nmi = metrics.normalized_mutual_info_score(df_gene["class"],clusters, average_method="geometric")
        nmis[k]=nmi
    max_k=max(nmis.items(), key=operator.itemgetter(1))[0]
    print("The maximum NMI with linkage={} is reached using {} clusters. The NMI value is {}.".format(l, max_k, round(nmis.get(max_k),2)))
```

As a result, Agglomerative clustering with ward-linkage and 6 clusters provides the best results, reaching NMI=0.92.

```python
labels=[]   
for k in [2, 6, 10]:
    clustering=AgglomerativeClustering(n_clusters=k, affinity='euclidean', compute_full_tree='auto', linkage="ward")
    values=clustering.fit_predict(df_gene_cleaned)
    labels.append(values)

pca = PCA(n_components=2)
principalComponents = pca.fit_transform(df_gene_cleaned)
fig, axes = plt.subplots(1, 3, figsize=(20,8))
axes[0].set_xlabel('Principal Component 1', fontsize = 15)
axes[0].set_ylabel('Principal Component 2', fontsize = 15)
axes[0].set_title('k=2', fontsize = 20)
axes[0].scatter(principalComponents[:,0], principalComponents[:,1],c=labels[0], s=50)
axes[0].grid()
axes[1].set_xlabel('Principal Component 1', fontsize = 15)
axes[1].set_ylabel('Principal Component 2', fontsize = 15)
axes[1].set_title('k=6', fontsize = 20)
axes[1].scatter(principalComponents[:,0], principalComponents[:,1],c=labels[1], s=50)
axes[1].grid()
axes[2].set_xlabel('Principal Component 1', fontsize = 15)
axes[2].set_ylabel('Principal Component 2', fontsize = 15)
axes[2].set_title('k=10', fontsize = 20)
axes[2].scatter(principalComponents[:,0], principalComponents[:,1],c=labels[2], s=50)
axes[2].grid()
```

### Ensemble method

```python
# Sort of ensemble method where Birch and Agglomerative clusterings are ran multiple times with different k-values, and the most common cluster is selected for each data point. 
def ensemble(data, pcs, k_list):
    labs=[]
    for pcnum in pcs:
        for k in k_list:
            pca = PCA(n_components=pcnum)
            pcs = pca.fit_transform(df_gene_cleaned)
            brc = Birch(n_clusters=k)
            brc.fit_predict(pcs)
            labs.append(brc.labels_)
            print(round(metrics.normalized_mutual_info_score(df_gene["class"],brc.labels_, average_method="geometric"),2))
            
            clustering = AgglomerativeClustering(n_clusters=k, affinity='euclidean', compute_full_tree='auto', linkage="ward")
            clusters=clustering.fit_predict(df_gene_cleaned)
            labs.append(clusters)
            print(round(metrics.normalized_mutual_info_score(df_gene["class"],clusters, average_method="geometric"),2))
    return np.array(labs)
```

```python
klist=[5,6,7]
pcs=[100,400,500]
cl_data=ensemble(df_gene_cleaned, pcs, klist)
```

```python
def most_common(lst):
    return max(set(lst), key=lst.count)

labels=[]
for i in range(0,795):
    vals=[]
    for val in range(0,18):
        vals.append(cl_data[val,i])
    labels.append(most_common(vals))
```

```python
round(metrics.normalized_mutual_info_score(df_gene["class"],labels, average_method="geometric"),2)
```

```python
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(df_gene_cleaned)
plt.scatter(principalComponents[:,0], principalComponents[:,1],c=labels, s=50)
plt.xlabel('Principal Component 1', fontsize = 15)
plt.ylabel('Principal Component 2', fontsize = 15)
plt.title('Ensemble method', fontsize = 20)
plt.grid()
```

Unfortunately, the used ensemble did not improve the performance. In general, the logic of the ensembling was probably way too simple, and more sophisticated algorithm would be needed for ensembling. 

### Optimizing the best performing method

```python
import random
features = [100, 300, 500, 800, 900, 1000, 1100, 1500, 2000, 3000, 4000]
K = [5,6]
K2 = [10, 50, 100, 1000]
pcnums = [2, 5, 10, 20, 50, 100, 200, 500, 700]
ths = [0.1, 0.3, 0.5, 0.7, 1.0, 2.0, 3.0]
bfs = [10, 50, 100]
score_function = [f_classif, chi2]

for i in range(0,100):
    feature=random.choice(features)
    k=random.choice(K)
    k2=random.choice(K2)
    pcnum=random.choice(pcnums)
    th=random.choice(ths)
    bf=random.choice(bfs)
    sf = random.choice(score_function)
    if pcnum > feature:
        continue
    results = []
    bestfeatures = SelectKBest(score_func=sf, k=k2)
    fit = bestfeatures.fit(df_gene_cleaned,df_gene["class"])
    dfscores = pd.DataFrame(fit.scores_)
    dfcolumns = pd.DataFrame(df_gene_cleaned.columns)
    featureScores = pd.concat([dfcolumns,dfscores],axis=1)
    featureScores.columns = ['Specs','Score']
    selected_data = df_gene_cleaned[featureScores.nlargest(feature, "Score").Specs]
    pca = PCA(n_components=pcnum)
    pcs = pca.fit_transform(selected_data)
    for j in range(0,20):
        brc = Birch(n_clusters=k, threshold=th, branching_factor=bf)
        brc.fit_predict(pcs)
        val = metrics.normalized_mutual_info_score(df_gene["class"],brc.labels_, average_method="geometric")
        results.append(val)
    avg_nmi = sum(results)/len(results)
    if avg_nmi > 0.94:
        print("Average NMI for {} clusters, {} features, {} pcas, {} threshold, {} branching factor, score_func {}, and {} best features: {}".format(k, feature, pcnum, th, bf, str(sf), k2, round((sum(results)/len(results)),4)))
```

### Final solution

```python
results = []
nmi_max=0

for i in range(0,20):
    # Select the best 1000 features
    bestfeatures = SelectKBest(score_func=f_classif, k=10)
    fit = bestfeatures.fit(df_gene_cleaned,df_gene["class"])
    dfscores = pd.DataFrame(fit.scores_)
    dfcolumns = pd.DataFrame(df_gene_cleaned.columns)
    featureScores = pd.concat([dfcolumns,dfscores],axis=1)
    featureScores.columns = ['Specs','Score']
    selected_data = df_gene_cleaned[featureScores.nlargest(1000, "Score").Specs]
    
    # Perform PCA for 100 components
    pca = PCA(n_components=100)
    pcs = pca.fit_transform(selected_data)
    
    # Perform Birch clustering to 5 clusters
    brc = Birch(n_clusters=5, threshold=1.0, branching_factor=10)
    brc.fit_predict(pcs)
    
    # Measure NMI
    nmi = metrics.normalized_mutual_info_score(df_gene["class"],brc.labels_, average_method="geometric")
    if(nmi>nmi_max):
        gene_labels=brc.labels_
    results.append(nmi)
    print("NMI: {}".format(round(nmi,3)))
print("GENE_DATA NMI average for 20 runs: {}".format(round(sum(results)/len(results),3)))
```

```python
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(df_gene_cleaned)
plt.scatter(principalComponents[:,0], principalComponents[:,1],c=gene_labels, s=50)
plt.xlabel('Principal Component 1', fontsize = 15)
plt.ylabel('Principal Component 2', fontsize = 15)
plt.title('Gene data clustering', fontsize = 20)
plt.show()
```
